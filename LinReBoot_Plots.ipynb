```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
from scipy import stats
import time
```

```python
weight_sd_list = [0.05, 0.1, 0.2, 0.5, 1]
color_list = ['b', 'c', 'g', 'k', 'm', 'y']
num_tuning = len(weight_sd_list)
horizon_len = 10000
# plots
plt.clf()
fig, (ax1, ax2, ax3) = plt.subplots(1,3)

########################################################################
## Experiment 1.1 tuning: Linear ReBoot-G Tuning
## setting: d = 5
## result: 
########################################################################

for i in range(num_tuning):
    file = "Results/SLB_res/setting_1/LinReBootG_res_" + str(i) + ".csv"
    label = "LinReBoot-G(" + str(weight_sd_list[i]) + ")"
    res = pd.read_csv(file)
    ax1.plot(np.arange(horizon_len), res['regret_avg'], label=label, color = color_list[i])
    ax1.fill_between(np.arange(horizon_len),
                    (res['regret_avg']-res['regret_std']),
                    (res['regret_avg']+res['regret_std']),
                    color = color_list[i], alpha = 0.2)

#ax1.legend(bbox_to_anchor=(0.33, 1))
ax1.set_xlabel("Decision Time", fontsize=22)
ax1.set_ylabel("Regret", fontsize=22)
ax1.tick_params(axis = "both", labelsize =20)
ax1.set_title("d=5", fontsize=30)


########################################################################
## Experiment 1.2 tuning: Linear ReBoot-G Tuning
## setting: d = 10
## result: 
########################################################################

for i in range(num_tuning):
    file = "Results/SLB_res/setting_2/LinReBootG_res_" + str(i) + ".csv"
    label = "LinReBoot-G(" + str(weight_sd_list[i]) + ")"
    res = pd.read_csv(file)
    ax2.plot(np.arange(horizon_len), res['regret_avg'], label=label, color = color_list[i])
    ax2.fill_between(np.arange(horizon_len),
                    (res['regret_avg']-res['regret_std']),
                    (res['regret_avg']+res['regret_std']),
                    color = color_list[i], alpha = 0.2)

#ax2.legend(bbox_to_anchor=(0.33, 1))
ax2.set_xlabel("Decision Time", fontsize=22)
ax2.set_ylabel("Regret", fontsize=22)
ax2.tick_params(axis = "both", labelsize =20)
ax2.set_title("d= 10", fontsize=30)


########################################################################
## Experiment 1.3 tuning: Linear ReBoot-G Tuning
## setting: d = 20
## result: 
########################################################################

for i in range(num_tuning):
    file = "Results/SLB_res/setting_3/LinReBootG_res_" + str(i) + ".csv"
    label = "LinReBoot-G(" + str(weight_sd_list[i]) + ")"
    res = pd.read_csv(file)
    ax3.plot(np.arange(horizon_len), res['regret_avg'], label=label, color = color_list[i])
    ax3.fill_between(np.arange(horizon_len),
                    (res['regret_avg']-res['regret_std']),
                    (res['regret_avg']+res['regret_std']),
                    color = color_list[i], alpha = 0.2)

ax3.legend(bbox_to_anchor=(0.8, -0.15), prop={'size': 22}, ncol=5)
ax3.set_xlabel("Decision Time", fontsize=22)
ax3.set_ylabel("Regret", fontsize=22)
ax3.tick_params(axis = "both", labelsize =20)
ax3.set_title("d=20", fontsize=30)

fig.set_size_inches(30, 8)
plt.show()
fig.savefig('Results/SLB_res/SLB_tuning_plot_summary.png', dpi=200,  bbox_inches='tight')
```
